# 查看数据分布

```python
import pandas as pd
import torch
from pathlib import Path
```

```python
data_dir = "../data/trec06c-utf8"
data_text_raw = Path(data_dir) / "data"
data_text_cut = Path(data_dir) / "data_cut"
label = pd.read_csv(Path(data_dir) /"label"/ "index", header=None, names=["label", "file_path"], sep=" ")
print(label.head())
```

```python
num_data_text_raw = sum(1 for p in data_text_raw.rglob("*") if p.is_file())
num_data_text_cut = sum(1 for p in data_text_cut.rglob("*") if p.is_file())
print(f"Number of raw text files: {num_data_text_raw}")
print(f"Number of cut text files: {num_data_text_cut}")
print(f"Number of labels: {len(label)}")
```
